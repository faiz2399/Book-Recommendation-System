```python
import pandas as pd
import numpy as np
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules
from math import nan, isnan
from mlxtend.frequent_patterns import apriori
import json
from urllib.request import urlopen
import time
```

```python
df = pd.read_excel('final_book.xlsx')
```

```python
df
```

```python
lst_book = df['ISBN'].to_list()
```

```python
lst_book = list(set(lst_book))
```

```python
from tqdm import tqdm
chunk_size = 300
lst_book1 = [lst_book[i:i + chunk_size] for i in range(0, len(lst_book), chunk_size)]
```

```python
isbn_book = []
des_book = []
api = "https://www.googleapis.com/books/v1/volumes?q=isbn:"
for book in tqdm(lst_book1[3]):
  isbn = book.strip()

  resp = urlopen(api + isbn)
  book_data = json.load(resp)
  if 'items' in book_data and 'volumeInfo' in book_data['items'][0]:
        isbn_book.append(book_data['items'][0]['volumeInfo']['industryIdentifiers'][0]['identifier'])
        if 'description' in book_data['items'][0]['volumeInfo']:
            des_book.append(book_data['items'][0]['volumeInfo']['description'])
        else:
            des_book.append("no description")
  else:
    isbn_book.append("no book")
    des_book.append("no category")
  time.sleep(5)

df_genre = pd.DataFrame({'ISBN':isbn_book,'Genre':des_book})
```
